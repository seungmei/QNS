## 2.2 양자 회로를 파이썬 프로그램으로 구현하기

### 2.2.1 양자 회로의 예 1


```python
from qiskit import QuantumCircuit

circuit = QuantumCircuit(3, 3) #3 qubit quantum register, 3 bit classical register

#양자레지스터의 상태를 0b111로 초기화
circuit.x(0)
circuit.x([2,1])

#회로에 배리어 표시
circuit.barrier()

circuit.ccx(2, 1, 0) #ccx(control=2, 1, target=0)

#측정 회로 추가
circuit.measure([0, 1, 2], [0, 1, 2])

circuit.draw(output='mpl', fold=-1) #fold=-1 : 게이트가 많은 회로를 한 줄에 모두 표현 가능.
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

qt = QuantumRegister(1, 'target') # 1 qubit QuantumRegister
qc = QuantumRegister(2, 'control') # 2 qubit QuantumRegister
cr = ClassicalRegister(3, 'm') # 3 bit ClassicalRegister
circuit = QuantumCircuit(qt, qc, cr) # 양자 회로 클래스의 객체를 생성

#양자레지스터의 상태를 0b111로 초기화
circuit.x(qt)
circuit.x(qc)

#회로에 배리어 표시
circuit.barrier()

circuit.ccx(qc[1], qc[0], qt)

#측정 회로 추가
circuit.measure([0, 1, 2], cr)

circuit.draw(output='mpl') #fold=-1 : 게이트가 많은 회로를 한 줄에 모두 표현 가능.
```

## 2.3 양자 회로를 실행하기

### 2.3.1 양자 회로를 시뮬레이션하기

```python
# AerSimulator를 이용하여 circuit 회로를 1000번 만들어 측정한 결과를 출력

from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit_aer import AerSimulator

aer_sim = AerSimulator() #시뮬레이션 객체 생성
pm = generate_preset_pass_manager(backend=aer_sim, optimization_level=1) # 가상의 양자 컴퓨터 디바이스를 위한 패스 매니저 객체 pm 생성
isa_circuit = pm.run(circuit) #양자 회로를 표현하는 circuit 객체를 양자 컴퓨터 디바이스에 구현할 수 있는 형태의 isa_circuit 객체로 transpile
sampler = Sampler(mode=aer_sim) #양자 회로를 시뮬레이션하고 출력 레지스터의 값들을 샘플링
job = sampler.run([isa_circuit], shots = 1000) #sampler 객체를 이용하여 isa_circuit을 aer_sim에서 1000회 시뮬레이션하는 job 시작
#isa_circuit은 회로 객체들의 list이다.

result = job.result() #job의 수행 결과를 리스트 타입 객체 result 에 치환
count = result[0].data.m.get_counts()
# result 리스트의 첫번째 원소: 첫번째 회로 isa_circuit의 시뮬레이션 결과
# 회로의 측정 결과인 고전 레지스터의 데이터를 표현하는 dict 타입의 객체 count 생성
# data의 멤버 변수명은 회로 다이어그램에 표시되는 고전 레지스터의 레이블과 같은 이름

print(count)
```

```python
from qiskit.visualization import plot_distribution
plot_distribution(count)
```

## 2.4 양자 회로의 모듈화(함수로 구현하기)


### 2.4.1 양자 회로를 함수로 만들기

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.barrier(qreg_q[0], qreg_q[1], qreg_q[2], qreg_q[3])
circuit.cx(qreg_q[0], qreg_q[2])
circuit.cx(qreg_q[1], qreg_q[2])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[3])
circuit.measure(qreg_q, creg_c)
# circuit.measure(qreg_q[0], creg_c[0])
# circuit.measure(qreg_q[1], creg_c[1])
# circuit.measure(qreg_q[2], creg_c[2])
# circuit.measure(qreg_q[3], creg_c[3])

circuit.draw(output='mpl')
```

```python
def half_adder(n):
  qr = QuantumRegister(n)
  circuit = QuantumCircuit(qr)
  circuit.cx(qr[0], qr[2]) # q0 q1 합을 q2에 저장
  circuit.cx(qr[1], qr[2])
  circuit.ccx(qr[0], qr[1], qr[3]) # q0 q1의 올림을 q3에 저장
  circuit.name = 'half_adder'
  return circuit
```

### 2.4.2 함수 호출하여 양자 회로 추가하기

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x([qreg_q[0], qreg_q[1]])
circuit.barrier()
c_ha = half_adder(4)
circuit.append(c_ha, [0, 1, 2, 3])
circuit.measure(qreg_q, creg_c)

circuit.draw(output='mpl')
```

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0:2])
circuit.barrier()
# circuit 회로 객체에 some_circuit을 추가
circuit.append(half_adder(4), [0, 1, 2, 3])
circuit.measure(qreg_q, creg_c)

circuit.draw(output='mpl')
```

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x(qreg_q[0:2])
circuit.barrier()
# .compose() 메소드는 inplace=False 가 기본값. => circuit 회로 객체는 변하지 않음. circuit 회로 객체에 some_circuit을 추가하여 생성된 새로운 회로 객체가 생성되어 반환
circuit.compose(half_adder(4), range(4), inplace=True)
circuit.measure(qreg_q, creg_c)

circuit.draw(output='mpl')
```

.append() 메소드로 추가한 회로와 다르게 큰 블록으로 표시 X
.compose() 는 개별 게이트들이 구체적으로 보여짐.

```python
qreg_q = QuantumRegister(8, 'q')
#creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q)

circuit.x([qreg_q[2], qreg_q[3]])
circuit.barrier()
# qubit 2, 3, 4, 6에 half_adder 회로를 추가하는 코드
circuit.compose(half_adder(4), qreg_q[2:4]+qreg_q[5:6]+qreg_q[7:8], inplace=True)

# 모든 큐비트를 측정하는 경우, 별도의 ClassicalRegister 객체를 선언할 필요가 없음.
# (이 함수는 자동으로 ClassicalRegister('meas') 객체를 생성하고, 모든 큐비트에 대한 측정장치를 포함)
circuit.measure_all()
circuit.draw(output='mpl')
```

```python
qreg_q = QuantumRegister(8, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x([qreg_q[2], qreg_q[3]])
circuit.barrier()
# qubit 2, 3, 4, 6에 half_adder 회로를 추가하는 코드
circuit.compose(half_adder(4), [2, 3, 5, 7], inplace=True)
circuit.measure([2,3,5,7],[0,1,2,3])
circuit.draw(output='mpl')
```

### 2.4.3 역회로Inverse Circuit 생성
- 모든 양자 연산은 유니터리 연산(=가역 연산)
- 모든 양자회로에는 그 역연산을 수행하는 역회로가 존재

```python
qreg_q = QuantumRegister(8, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.x([qreg_q[2], qreg_q[3]])
circuit.barrier()
c_ha = half_adder(4)
circuit.compose(c_ha, range(2,6), inplace=True)
circuit.barrier()
# 반가산기의 역회로를 생성하여 추가하는 코드
circuit.compose(c_ha.inverse(), range(2,6), inplace=True)
circuit.barrier()
circuit.measure(range(2,6), creg_c)
circuit.draw(output='mpl')
```

### 2.4.4 다중 제어 X 게이트의 제어 상태 설정
- QuantumCircuit.mcx() 함수는 다중제어 X 게이트를 구현하는 데 사용되는 함수.
  - 제어큐비트가 3개 이상일 때 사용.
  - ( 제어큐비트 리스트, 타겟 큐비트)
  - 제어 큐비트가 모두 1> 일 때 X gate 적용.

```python
def mcx_test(controls, targets):
  qc = QuantumCircuit(controls, targets)
  #ctrl_state = '0': controls[0] qubit가 0> 상태일 때, targets[0] 에 X gate를 적용
  qc.cx(controls[0], targets[0], ctrl_state='0')
  #ctrl_state = '01': controls[0]이 1> 상태이고, controls[1]이 0> 상태일 때, targets[1] 큐비트에 X gate 연산 적용
  qc.ccx(controls[0], controls[1], targets[1], ctrl_state='01')
  #ctrl_state를 명시하지 않았기 때문에, 모두 111 상태일 때 X 게이트 적용
  qc.mcx([controls[0], controls[1], controls[2], controls[3]], targets[2])
  qc.mcx(controls[0:4], targets[3], ctrl_state='0001')
  qc.name = 'mcx_test'
  return qc

qreg_c = QuantumRegister(4, 'controls')
qreg_t = QuantumRegister(4, 'targets')
qreg_etc = QuantumRegister(2, 'etc')
creg = ClassicalRegister(4, 'measure')
circuit = QuantumCircuit(qreg_etc, qreg_c, qreg_t, creg)

circuit.compose(mcx_test(qreg_c, qreg_t), qreg_c[0:4]+qreg_t[:4], inplace=True)
circuit.barrier()
circuit.compose(mcx_test(qreg_c, qreg_t).inverse(), range(2,10), inplace=True)
circuit.measure(qreg_t, range(3, -1, -1)) #[3, 2, 1, 0]
circuit.draw(output='mpl')
```
